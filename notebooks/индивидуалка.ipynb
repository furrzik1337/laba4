```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  Series         
dates = pd.date_range('2024-06-01', periods=30, freq='D')
data = np.random.randint(0, 101, 30)  #   0  100
series = pd.Series(data, index=dates)

#  ,    80
high_popularity_dates = series[series > 80].index

#  
plt.figure(figsize=(12, 6))
plt.plot(series, label='')

#    ,    80
plt.plot(high_popularity_dates, series[high_popularity_dates], 'ro', label=' > 80') # 'ro' -  

#     
plt.title('    30 ', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)

#  
plt.grid(True)

#    x   
plt.xticks(rotation=45, ha='right')

#  
plt.legend()

#  
plt.tight_layout()
plt.show()
```
